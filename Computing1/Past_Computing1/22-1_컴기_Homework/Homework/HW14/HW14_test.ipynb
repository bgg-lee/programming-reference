# P1


```python
def P1(n: int) -> bool:
    # write your code below
    lst = []
    bn = bin(n)
    ans = True
    curr = bn[2]
#     print(bn)
    
    for i in bn[3:]:
        if i == curr:
            ans = False
            curr = i
#             print(ans)
        else:
            curr = i
#             print(ans)
  
    return ans
```

# P2


```python
def P2(num:int)->int:
    # write your code below
    bn = bin(num)[2:]
#     print(bn, len(bn))  
    bn = bn.zfill(32)    
#     print(bn, len(bn))
    ans = bn[::-1]
#     print(ans, len(ans))
       
    return int(ans, 2)
```

# P3


```python
def P3(nums:list)->set:
    # write your code below
    ans = []
    dct = {}
    
    for i in nums:
        if i in dct:
            dct[i] +=1
        else:
            dct[i] = 1
#     print(dct)
    
    for i,j in dct.items():
        if j == 1:
            ans.append(i) 
#     print(ans)
  
    return set(ans)
```

```python
P3([1,2,1,3,2,5])
```

```python
P3([-1,0])
```

```python
P3([1,2,3,4,5,6,7,8,9,1,2,3,4,5,6,7])
```

# P4


```python
def P4(num:int) -> int:
    # write your code below
    ans = 0
    bn = bin(num)[2:]
#     print(bn)
    
    for i in bn:
        if i == '1':
            ans += 1

    return ans
```

# P5


```python
P5_input1 = ['a', 'B', 'c', 'D']
P5_input2 = ['z', 'b', 'k', 'D', 'I', 'q']
P5_input3 = ['a', 'b']
```

```python
from typing import List
def P5(alphabet_list: List[str]) -> int:
    idx = 0
    ### Modify code here ###
    
    while idx < len(alphabet_list):
        if alphabet_list[idx].islower():
            idx += 1
        elif alphabet_list[idx].isupper():
            break
     
    else: 
        idx = -1
    
    ### End of your code ###  
    return idx
```

```python
P5(P5_input1)
```

```python
P5(P5_input2)
```

```python
P5(P5_input3)
```

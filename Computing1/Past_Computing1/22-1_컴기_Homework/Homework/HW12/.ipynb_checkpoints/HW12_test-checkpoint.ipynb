```python
def P1(nums: list, k: int) -> bool:

    ### Write code here ###
    dct = dict()
    
    for  i in nums:
        if i% k  in dct:
            dct[i%k] += 1
        else:
            dct[i%k] = 1
    
    for i in dct:
        if i == 0 or i * 2 == k:
            if dct[i] % 2 != 0:
                return False
#             else:
#                 return True
        else:
            if k-i not in dct:
                return False
            if dct[i] != dct[k-i]: 
                return False
    return True
    
    

    
    ### End of your code ###
```

```python
def P2(nums: list) -> int:

    ### Write code here ###
    
    ans = 0
    now = 0
    sum_idx = dict()
    
    for i in range(len(nums)):
        if nums[i] == 0:
            now -= 1
        elif nums[i] == 1:
            now += 1
        
        if now == 0:
            ans = i+1
        else:
            if now in sum_idx:
                ans = max(ans, i- sum_idx[now])
            else:
                sum_idx[now] = i
    
    return ans
    
    ### End of your code ###  
```

```python
def P4(nums: list) -> int:

    ### Write code here ###
    #   
       
    len_max = 0        
    
    def find_conti(nums: list, num: int, conti: int) -> int:
        for i in range(len(nums)):
            if nums[i] == num+1:
                num += 1
                conti += 1
                return find_conti(nums,num,conti)
        return conti
        
    for i in range(len(nums)):
        conti = find_conti(nums, nums[i], 1)  # conti
        if len_max < conti:
            len_max = conti
#             max_start = nums[i]s
        
    return len_max
    
    
    
    ### End of your code ###  
```

```python
def P4(nums: list) -> int:

    ### Write code here ###
    nums = set(nums)
    dic = {}
    dic_rev = {}
    
    for i in nums:
        if (i+1 not in dic) and (i-1 not in dic_rev):
            dic[i] = i  # {3:3, 7:7}
            dic_rev[i] = i  # {3:3, 7:7}
            continue
            
        if i+1 in dic:
            dic[i] = dic[i+1]  # {3:4, 4:4}
            dic_rev[dic[i+1]] = i  # {4:4}
            del dic[i+1] # {3:4}
            
        if i-1 in dic_rev:
            dic_rev[i] = dic_rev[i-1]
            dic[dic_rev[i-1]] = i
            del dic_rev[i-1]
            
        if (i in dic) and (i in dic_rev):
            dic[dic_rev[i]] = dic[i]
            dic_rev[dic[i]] = dic_rev[i]
            del dic[i]
            del dic_rev[i]
    
    len_max = 0    
    for i in dic:
        if len_max < dic[i] - i + 1:
            len_max = dic[i] - i + 1
                        
    return len_max   
    
    ### End of your code ###  
```

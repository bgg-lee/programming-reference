Given an array of distinct integers candidates and a target integer target, return a list of all unique combinations of candidates where the chosen numbers sum to target. You may return the combinations in any order.

The same number may be chosen from candidates an unlimited number of times. Two combinations are unique if the 
frequency
 of at least one of the chosen numbers is different.

The test cases are generated such that the number of unique combinations that sum up to target is less than 150 combinations for the given input.

 

Example 1:

Input: candidates = [2,3,6,7], target = 7
Output: [[2,2,3],[7]]
Explanation:
2 and 3 are candidates, and 2 + 2 + 3 = 7. Note that 2 can be used multiple times.
7 is a candidate, and 7 = 7.
These are the only two combinations.
Example 2:

Input: candidates = [2,3,5], target = 8
Output: [[2,2,2,2],[2,3,3],[3,5]]
Example 3:

Input: candidates = [2], target = 1
Output: []
 

Constraints:

1 <= candidates.length <= 30
2 <= candidates[i] <= 40
All elements of candidates are distinct.
1 <= target <= 40

candidate list에서 sum을 만들 수 있는 모든 조합을 반환하는 것
같은 수가 여러 번 쓰일 수도 있다.

```python
# 같은 놈
from typing import List 
class Solution:
    def combinationSum(self, candidates: List[int], target: int) -> List[List[int]]:
        def backtrack(idx,curr,lst):
            if curr == target:
                ans.append(lst[:])
                return
            elif curr > target : return

            for i in range(idx,len(candidates)):
                lst.append(candidates[i])
                backtrack(i,curr+candidates[i],lst)
                lst.pop()
        ans = []
        backtrack(0,0,[])
        return ans
s = Solution()
s.combinationSum([5,7,3],13)
```

```python
# C -> 할당 미쳤다
#include <stdio.h>
#include <stdlib.h>

// 백트래킹 함수
void backtrack(int* candidates, int candidatesSize, int target, int start, 
               int* curr, int currSize, int currSum, 
               int*** result, int* returnSize, int** returnColumnSizes) {
    if (currSum == target) {
        // 새로운 조합을 저장할 공간 할당
        (*result)[*returnSize] = (int*)malloc(currSize * sizeof(int));
        for (int i = 0; i < currSize; i++) {
            (*result)[*returnSize][i] = curr[i];
        }
        (*returnColumnSizes)[*returnSize] = currSize;
        (*returnSize)++;
        return;
    }
    if (currSum > target) return;

    for (int i = start; i < candidatesSize; i++) {
        curr[currSize] = candidates[i];
        backtrack(candidates, candidatesSize, target, i, curr, currSize + 1, 
                  currSum + candidates[i], result, returnSize, returnColumnSizes);
    }
}

// 조합 합 함수
int** combinationSum(int* candidates, int candidatesSize, int target, 
                     int* returnSize, int** returnColumnSizes) {
    *returnSize = 0;
    int maxCombinations = 500;  // 충분히 큰 배열 크기 설정
    int** result = (int**)malloc(maxCombinations * sizeof(int*));
    *returnColumnSizes = (int*)malloc(maxCombinations * sizeof(int));

    int* curr = (int*)malloc(target * sizeof(int));  // 최대 target 개수까지 저장 가능
    backtrack(candidates, candidatesSize, target, 0, curr, 0, 0, 
              &result, returnSize, returnColumnSizes);

    free(curr);
    return result;
}
```

```python
# C++ Stack으로 풀기
# C++ Stack , vector 등 다양한 container에 다양한 data type의 data를 집어넣으려면 tuple을 이용해야한다.
class Solution {
public:
    vector<vector<int>> combinationSum(vector<int>& candidates, int target) {
        // start idx, curr sum, temp vector
        stack<tuple<int,int,vector<int>>> s;
        vector<vector<int>> ans;
        s.push({0,0,{}});
        while(!s.empty()){
            auto [idx, curr, vec] = s.top();
            s.pop();
            if(curr==target){
                ans.push_back(vec);
                continue;
            }
            else if(curr>target) continue;

            for(int i=idx;i<candidates.size();i++){
                vector<int> temp = vec;
                temp.push_back(candidates[i]); # 이런식으로 추가하면 된다.
                s.push({i,curr+candidates[i],temp});
            }
        }
        return ans;
    }
};
```

std::tuple<double, char, std::string> get_student(int id)

        case 2: return {1.7, 'D', "Ralph Wiggum"};

    throw std::invalid_argument("id");
}

You are given a 0-indexed integer array arr, and an m x n integer matrix mat. arr and mat both contain all the integers in the range [1, m * n].

Go through each index i in arr starting from index 0 and paint the cell in mat containing the integer arr[i].

Return the smallest index i at which either a row or a column will be completely painted in mat.

 

Example 1:

image explanation for example 1
Input: arr = [1,3,4,2], mat = [[1,4],[2,3]]
Output: 2
Explanation: The moves are shown in order, and both the first row and second column of the matrix become fully painted at arr[2].
Example 2:

image explanation for example 2
Input: arr = [2,8,7,4,1,3,5,6,9], mat = [[3,2,5],[1,4,6],[8,7,9]]
Output: 3
Explanation: The second column becomes fully painted at arr[3].
 

Constraints:

m == mat.length
n = mat[i].length
arr.length == m * n
1 <= m, n <= 105
1 <= m * n <= 105
1 <= arr[i], mat[r][c] <= m * n
All the integers of arr are unique.
All the integers of mat are unique.

```python
# 1 Bingo 빙고 찾으면 끝나는 문제
# 잘 한 거 같은데 내 풀이.. -_- time exceed
class Solution:
    def firstCompleteIndex(self, arr: List[int], mat: List[List[int]]) -> int:
        m = len(mat)
        n = len(mat[0])
        # m * n Matrix
        visited = [ [False] * n for _ in range(m) ]

        def bingo(temp:list[list[int]]) -> bool :
            # row check
            for row in range(m):
                if False not in temp[row]:
                    return True
            # column check
            for col in range(n):
                row_col = []
                for row in range(m):
                    row_col.append(temp[row][col])
                if False not in row_col :
                    return True
            return False

        for val in arr:
            for i in range(m):
                for j in range(n):
                    if mat[i][j] == val :
                        visited[i][j] = True
                        break
            if bingo(visited): return arr.index(val)

```

```python
# dictionary로 시간효율 개선한 버전, 엄청난 센스다
class Solution:
    def firstCompleteIndex(self, arr: List[int], mat: List[List[int]]) -> int:
        m, n = len(mat), len(mat[0])
        # Map values in mat to their (row, column) indices
        val_to_pos = {mat[i][j]: (i, j) for i in range(m) for j in range(n)}

        # Arrays to track painted cells in each row and column
        row_count = [0] * m
        col_count = [0] * n

        for i, val in enumerate(arr):
            # Find the position of the current value in mat
            row, col = val_to_pos[val]

            # Update counts for the corresponding row and column
            row_count[row] += 1
            col_count[col] += 1

            # Check if the row or column is completely painted
            if row_count[row] == n or col_count[col] == m:
                return i

        return -1  # This should never be reached given the problem constraints

"""
위와 같은 풀이이
class Solution:
    def firstCompleteIndex(self, arr: List[int], mat: List[List[int]]) -> int:
        m, n = len(mat),len(mat[0])
        mat_val = {mat[i][j]:(i,j) for i in range(m) for j in range(n)}

        row_cnt = [0] * m
        col_cnt = [0] * n

        for i in range(len(arr)):
            row, col = mat_val[arr[i]]
            
            row_cnt[row] += 1
            col_cnt[col] += 1

            if row_cnt[row] == n or col_cnt[col] == m : return i
"""
```

```python
"""
C 풀이
#include <stdio.h>
#include <stdlib.h>
#include <stdbool.h>

// 좌표 구조체 정의
typedef struct {
    int row;
    int col;
} Position;

int firstCompleteIndex(int* arr, int arrSize, int** mat, int matSize, int* matColSize) {
    int m = matSize;
    int n = matColSize[0]; // 모든 열의 크기가 동일하다고 가정

    // 값을 좌표로 매핑할 배열 (1부터 m * n까지 매핑)
    Position* val_to_pos = (Position*)malloc((m * n + 1) * sizeof(Position));
    for (int i = 0; i < m; i++) {
        for (int j = 0; j < n; j++) {
            val_to_pos[mat[i][j]] = (Position){i, j};
        }
    }

    // 행과 열의 카운트를 저장할 배열
    int* row_count = (int*)calloc(m, sizeof(int));
    int* col_count = (int*)calloc(n, sizeof(int));

    // 배열을 순회하며 행 또는 열이 모두 칠해졌는지 확인
    for (int i = 0; i < arrSize; i++) {
        int val = arr[i];
        Position pos = val_to_pos[val];
        int row = pos.row;
        int col = pos.col;

        // 행과 열의 카운트를 업데이트
        row_count[row]++;
        col_count[col]++;

        // 행 또는 열이 완전히 칠해졌는지 확인
        if (row_count[row] == n || col_count[col] == m) {
            free(val_to_pos);
            free(row_count);
            free(col_count);
            return i; // 인덱스 반환
        }
    }

    // 메모리 해제
    free(val_to_pos);
    free(row_count);
    free(col_count);
    return -1; // 기본 반환값
}

"""
```

```python
# 역시나 time limit 걸리는 내 C++ 풀이 -_-
class Solution {
public:
    bool bingo(const vector<vector<bool>>& temp){
        int m = temp.size();
        int n = temp[0].size();
        // row check
        for(int i=0;i<m;i++){
            if(find(temp[i].begin(),temp[i].end(),false) == temp[i].end()) return true;
        }
        // col check
        for(int j=0;j<n;j++){
            vector<bool> arr = {};
            for(int i=0;i<m;i++){
                arr.push_back(temp[i][j]);
            }
            if(find(arr.begin(),arr.end(),false) == arr.end()) return true;
        }
        // No bingo case
        return false;
    }

    int firstCompleteIndex(vector<int>& arr, vector<vector<int>>& mat) {
        int m = mat.size();
        int n = mat[0].size();
        vector<vector<bool>> visited(m,vector(n,false));

        for(int k=0;k<arr.size();k++){
            bool cond = false;
            for(int i=0;i<m;i++){
                for(int j=0;j<n;j++){
                    if(mat[i][j]==arr[k]){
                        visited[i][j] = true;
                        cond = true;
                        break;
                    }
                }
                if(cond) break;
            }
            if(bingo(visited)) return k;
        }

        return -1;
    }
};
```
